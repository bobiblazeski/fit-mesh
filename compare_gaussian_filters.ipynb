```python
from src.discrete_gaussian import DiscreteGaussian
from src.operators import get_gaussian

import numpy as np
import seaborn as sns
import matplotlib.pylab as plt

import torch
np.set_printoptions(precision=4, suppress=True)

torch.set_printoptions(precision=4, sci_mode=False)
```

```python
size, sigma = 5, 1
#n= np.zeros((size,size))
#n[(size-1) // 2, (size-1) // 2] = 1
k1 = DiscreteGaussian(size,sigma=sigma)
k1 = k1.seq[1].weight.data[0, 0]
print(k1.shape, k1.sum())

ax = sns.heatmap(k1, linewidth=0.5)
plt.show()


k2 = get_gaussian(size)
k2 = k2.weight.data[0, 0]
print(k2.shape, k2.sum())

ax = sns.heatmap(k2, linewidth=0.5)
plt.show()

print((k1 -k2).sum())
ax = sns.heatmap(k1 -k2, linewidth=0.5)

plt.show()
```
